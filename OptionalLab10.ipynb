```python
import numpy as np
import matplotlib.pyplot as plt
import advection_funs
from importlib import reload
%matplotlib inline
```

```python
## order one polynomial
timesteps = 24*3600*20
cmatrix = advection_funs.advection3(timesteps,1)
ana_matrix = advection_funs.analytical('1',timesteps)
fig, axs = plt.subplots(1,2,figsize = (15,6))
axs[0].plot(ana_matrix[0], color = 'Olive', marker = '*',label = 'analytical')
axs[0].plot(cmatrix[-1,:], color = 'tomato', marker = 'o',label = 'numerical')
axs[1].plot(ana_matrix[0] - cmatrix[-1,:], color = 'g', label = 'difference')
for ax in axs:
    ax.legend(loc = 'best')
print ('total absolute error:%s'%np.sum(np.abs(cmatrix[-1,:]-ana_matrix[0])))
```
