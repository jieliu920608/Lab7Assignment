```python
import matplotlib.pyplot as plt
import numpy as np
import sys
from importlib import reload
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
```

```python
import interactive1 
```

```python
H0 = 1000;
L = 1000;
ngrid = 11
X = np.arange(0,L+100,100)
Y = np.arange(0,L+100,100)
X, Y = np.meshgrid(X, Y)
```

```python
## With flat bottom
interactive1.interactive1(3, 11, 150, 4*3600, small = False, flag ='flat')
Hu, Hv = interactive1.find_depth3(H0, ngrid, flag ='flat')
Z = Hu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, alpha=0.3,color = 'Olive')
cset = ax.contour(X, Y, Z, zdir='z', offset=Z.max(), cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=X.max(), cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=Y.max(), cmap=cm.coolwarm)
ax.view_init(azim=260)
ax.set_title('Bottom topography')
ax.set_xlabel('x [km]')
ax.set_ylabel('y [km]')
ax.set_zlabel('z [m]')
ax.invert_zaxis()
```

However, from streamfunction plot, magnitudes of both u and v increase around center of the domain with the Gaussian bottom topography compared with flat topography, which should be due to the conservation of mass. That is, with compression of the center of domain in the Gaussian bottom case, u and v velocity should increase correspondingly to satisfy mass conservation. Besides, the surface height of Gaussian bottom varied much compared with that of flat bottom, which is due to more change of depth in Gaussian topography, so from the formula $\frac{\partial h}{\partial t} + \nabla_{h} \cdot h\vec{u} = 0$, with more variation of second term, the first term should also has more variation amplitude of height.
